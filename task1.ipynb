# categorical features here are Sex, Pclass, and Embarked. One hot encoding technique will be used to to represent categorical variables as numerical values for our machine learning model.

```python
sex=pd.get_dummies(titanic_data['Sex'], drop_first=True)
sex.head(5)
```

```python
embarked = pd.get_dummies(titanic_data['Embarked'], drop_first=True)
embarked.head(5)
```

```python
Pcl=pd.get_dummies(titanic_data['Pclass'], drop_first=True)
Pcl.head(5)
```

```python
titanic_data = pd.concat([titanic_data, sex, embarked,Pcl], axis=1)
```

```python
titanic_data.head(5)
```

```python
titanic_data.drop(['Name', 'PassengerId', 'Pclass', 'Ticket', 'Sex', 'Embarked'], axis=1, inplace=True)
```

```python
titanic_data.head(5)
```

```python
X=titanic_data.drop('Survived', axis=1)
y=titanic_data['Survived']
```

```python
X.columns = X.columns.astype(str)
```

```python
from sklearn.model_selection import train_test_split
```

```python
pip install scikit-learn
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.33, random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
lm=LogisticRegression(max_iter=150)
```

```python
lm.fit(X_train,y_train)
```

```python
predictions= lm.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#checking accuracy of model
accuracy_score(y_test,predictions)
```
